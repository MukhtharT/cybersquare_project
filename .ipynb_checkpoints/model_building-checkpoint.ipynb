```python
categorical_cols=['location','find_us','qualification','branch','career_issue','looking_for','technology','attended_by','remarks']
data = pd.get_dummies(data, columns=categorical_cols,dtype=int)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
data['mode']=le.fit_transform(data['mode'])
# online 2
# ofline 1
# hybrid 0
```

```python
data['mode']
```

```python
data['time_slot']=data['time_slot'].map({'morning':3,'evening':2,'afternoon':1,'weekend':0})
```

```python
data['time_slot']
```

```python
data['status']=data['status'].map({'joined':1,'not joined':0})
```
